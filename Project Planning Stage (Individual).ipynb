Compute the mean value for each quantitative variable in the players.csv data set. Report the mean values in a table format.

```python
players_means <- summarise(players,
  mean_age = round(mean(Age, na.rm = TRUE), 2),
  mean_hours = round(mean(played_hours, na.rm = TRUE), 2)
)
```

Third step. A histogram — to help to see how playing time differs between subscribers and non-subscribers.

```python
players |>
  ggplot(aes(x = played_hours, fill = subscribe)) +
  geom_histogram(position = "dodge") +
  labs(title="Play Hours by Subscription", x="Hours Played", y="Count")
```
