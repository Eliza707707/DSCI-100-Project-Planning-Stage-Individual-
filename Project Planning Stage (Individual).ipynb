Third step. A histogram — to help to see how playing time differs between subscribers and non-subscribers.

```python
players |>
  ggplot(aes(x = played_hours, fill = subscribe)) +
  geom_histogram(position = "dodge") +
  labs(title="Play Hours by Subscription", x="Hours Played", y="Count")
```

(4) Methods and Plan
Method: I will use logistic regression to predict whether a player subscribes or not.
Why this method: It’s an easy and clear way to see how different factors (like play hours or experience) affect a yes/no outcome.
What it assumes: Each player’s data is separate, and the link between playtime and the chance of subscribing is fairly straight.
Weak points: It might miss more complex patterns (for example, if too much playtime stops helping) or other reasons why people subscribe that we can’t see in the data.
How I’ll check if it works: I’ll split the data — use 75% to train the model and 25% to test how well it predicts. I’ll look at how accurate it is overall.
Steps:
