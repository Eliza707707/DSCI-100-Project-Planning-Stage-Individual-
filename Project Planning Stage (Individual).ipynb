(1) Data Description
Files: players.csv (196 rows × 7 vars) and sessions.csv (1,535 rows × 5 vars).
players.csv variables:
hashedEmail (key to join with sessions), name (string), gender (category), experience (category), subscribe (logical), played_hours (numeric), Age (numeric; 2 missing).
sessions.csv variables:
hashedEmail, start_time, end_time (strings in dd/mm/yyyy HH:MM), and original_* timestamps (epoch ms).
Potential issues:
Time parsing, possible session overlaps or zero/negative durations, missing Age, self-report/measurement bias, duplicate players by name, class imbalance (e.g., few subscribers or rare play windows).
